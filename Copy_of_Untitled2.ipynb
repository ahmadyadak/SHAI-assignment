```python

ds_salaries.drop_duplicates(inplace=True)
```

```python
ds_salaries.duplicated().sum()
```

```python
company_size_change = ds_salaries['company_size'].replace({'S': 0, 'M': 1, 'L': 2})
# S = 0
# M = 1
# L = 2
ds_salaries['company_size'] = company_size_change
ds_salaries.head()
```

```python
experience_level_change = ds_salaries['experience_level'].replace({'SE': 0, 'MI': 1, 'EN': 2, 'EX': 3})
# SE = 0
# MI = 1
# EN = 2
# EX = 3
ds_salaries['experience_level'] = experience_level_change
ds_salaries.head()
```

```python
ds_salaries['work_year'] = ds_salaries['work_year'].replace(2020, 20)
ds_salaries['work_year'] = ds_salaries['work_year'].replace(2021, 21)
ds_salaries.head()
```

```python
#ds_salaries = ds_salaries.drop('salary_currency', axis=1)
#ds_salaries = ds_salaries.drop('salary', axis=1)
ds_salaries = ds_salaries.drop('un', axis=1)

ds_salaries.head()
```

```python

employee_residence_change = ds_salaries['employee_residence'].replace({'DE': 0, 'JP' :1, 'GB': 2, 'US': 3, 'CA': 4, 'IN': 5, 'FR': 6, 'ES': 7, 'GR': 8, 'TR': 9, 'NL': 10, 'BR': 11, 'PT': 12, 'PL': 13, 'IT': 14, 'RU': 15, 'PK': 16, 'AU': 17, 'AE': 18, 'MX': 19, 'CH': 20, 'NG': 21, 'BE': 22, 'SI': 23, 'DK': 24, 'AT': 25, 'HU': 26, 'RO': 27, 'CZ': 28, 'IE': 29, 'VN': 30, 'CL': 31, 'SG': 32, 'CO': 33, 'MY': 34, 'AR': 35, 'UA': 36, 'IR': 37, 'MT': 38, 'KE': 39, 'CN': 40, 'HN': 41, 'DZ': 42, 'MD': 43, 'LU': 44})
# DE = 0
# JP = 1
# GB = 2
ds_salaries['employee_residence'] = employee_residence_change
ds_salaries.head()
```

```python
ds_salaries['employment_type'] = ds_salaries['employment_type'].replace('FT', 0)
ds_salaries.head()
```
